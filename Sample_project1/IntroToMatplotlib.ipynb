## Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for jupyter
%matplotlib inline
import matplotlib.pyplot as plt
# 1. Prepare data
x = [1,2,3,4]
y= [11,22,33,44]
# 2. Setup plot
fig , ax = plt.subplots(figsize = (10,10)) # (width , hight)
# 3. Plot data
ax.plot(x,y)
# 4. Customize plot
ax.set(
    title=  "Simple Plot",
    xlabel= "x-axis",
    ylabel= "y-axis"
)
# 5. Save & show (you save the whole figure)
fig.savefig("figures/sample-plot.png")
```

```python
# Create some data
x = np.linspace(start = 0, stop= 10, num = 100)
```

```python
#Make a plot from a dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter":8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),height= nut_butter_prices.values())
ax.set(
    title=  "Dan's nut butter store",
    xlabel= "Butter kind",
    ylabel= "Price ($)"
)
plt.show()
```

```python
# Horizontal Bar plot
fig,ax = plt.subplots()
#need to convert to lists
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
plt.show()
```

```python
# Histogram
np.random.seed(0)
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x)
plt.show()
```

## Two options for subplots

Subplots option 1:

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows= 2,
                                          ncols= 2,
                                          figsize = (10,5))
#Plotting different axes
x1 = x
y1 = x/2

np.random.seed(0)
x2 = np.random.random(size = 10)
np.random.seed(1)
y2 = np.random.random(size = 10)

x3 = nut_butter_prices.keys()
y3 = nut_butter_prices.values()

np.random.seed(2)
y4 = np.random.randn(1000)

ax1.plot(x,y1)
ax2.scatter(x2,y2)
ax3.bar(x3,y3)
ax4.hist(y4)
plt.show()
```

Subplots option 2

```python
fig ,ax = plt.subplots(nrows= 2,
                       ncols=2,
                       figsize = (10,5))
#Plotting different axes by *Index*
x1 = x
y1 = x/2

np.random.seed(0)
x2 = np.random.random(size = 10)
np.random.seed(1)
y2 = np.random.random(size = 10)

x3 = nut_butter_prices.keys()
y3 = nut_butter_prices.values()

np.random.seed(2)
y4 = np.random.randn(1000)


ax[0,0].plot(x1,y1)
ax[0,1].scatter(x2,y2)
ax[1,0].bar(x3,y3)
ax[1,1].hist(y4)
plt.show()
```

## Example : Heart disease dataset

```python
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease.head()
```

- Age distribution

```python
heart_disease["age"].plot.hist(bins = 50)
plt.show()
```

 - Subplots: BAD

```python
heart_disease.plot.hist(subplots = True,figsize = (10,40),bins= 20)
plt.show()
```

 - Subplots: Better

```python
# Subsetting for patients over 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind = "scatter",
             x = "age",
             y = "chol",
             c= "target")
plt.show()
```

```python
# OO method with pyplot
fig,ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c= "target",ax = ax)
# ax.set_xlim([50 -5,100])
plt.show()
```

## OO method from scratch

```python
fig, ax = plt.subplots(figsize = (10,6))

#Plot data
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"])
# Customize
ax.set(
    title= "Heart disease and Cholesterol Levels",
    xlabel= "Age",
    ylabel= "Cholesterol"
)
# Adding a legend
ax.legend(*scatter.legend_elements(),title = "Target")
# Adding a mean line
ax.axhline(xmin = 0,xmax= 1,
           y = over_50["chol"].mean(),
           linestyle = ":",
           linewidth = 3)
plt.show()
```

```python
# Adding  a subplot for age vs thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize = (10,10),
                              sharex= True)
#Plotting data
# Ax0 ----------------------
scatter_plot0 = ax0.scatter( 
                     x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])
ax0.set(title= "Heart disease and Cholesterol Levels",
        # x label shared
        ylabel= "Cholesterol")
ax0.legend(*scatter_plot0.legend_elements(), title = "Target")
ax0.axhline( y = over_50["chol"].mean(),
           linestyle = ":",
           linewidth = 3)

# Ax1 ----------------------
scatter_plot1 = ax1.scatter(x = over_50["age"],
                     y = over_50["thalach"],
                     c = over_50["target"])
ax1.set(title= "Heart disease and Max heart rate Levels",
        xlabel= "Age",
        ylabel= "Max heart rate (Thalach)")
ax1.legend(*scatter_plot1.legend_elements(), title  = "Target")

ax1.axhline( y = over_50["thalach"].mean(),
            linestyle = "--",
            linewidth = 3)

# Figure customization
fig.suptitle(t= "Heart Disease Analysis", 
            fontsize = 20,
            fontweight = "bold")
plt.show()
```

## Altering the styles
 * Changing style from an existing style

```python
# Set the style
plt.style.use("seaborn-v0_8-whitegrid")
# Add color map to scatter_plot later
# ----------------------
fig, ax = plt.subplots(figsize = (10,6))

#Plot data
scatter_plot = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"],
                    cmap= "winter")
# Customize
ax.set(
    title= "Heart disease and Cholesterol Levels",
    xlabel= "Age",
    ylabel= "Cholesterol"
)
# Adding a legend
ax.legend(*scatter_plot.legend_elements(),title = "Target")
# Adding a mean line
ax.axhline(xmin = 0,xmax= 1,
           y = over_50["chol"].mean(),
           linestyle = ":",
           linewidth = 3)
plt.show()
```

## Another example

```python
# Customizing the y and x axis bounds

# Adding  a subplot for age vs thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize = (10,10),
                              sharex= True)
#Plotting data
# Ax0 ----------------------
scatter_plot0 = ax0.scatter( 
                     x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],
                     cmap = "winter")
ax0.set(title= "Heart disease and Cholesterol Levels",
        # x label shared
        ylabel= "Cholesterol")
# Change axes limits
ax0.set_xlim([50,80])
#  -------------------------
ax0.legend(*scatter_plot0.legend_elements(), title = "Target")
ax0.axhline( y = over_50["chol"].mean(),
           linestyle = ":",
           linewidth = 3)

# Ax1 ----------------------
scatter_plot1 = ax1.scatter(x = over_50["age"],
                     y = over_50["thalach"],
                     c = over_50["target"],
                     cmap = "winter")
ax1.set(title= "Heart disease and Max heart rate Levels",
        xlabel= "Age",
        ylabel= "Max heart rate (Thalach)")
# Change axes limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#  -------------------------
ax1.legend(*scatter_plot1.legend_elements(), title  = "Target")

ax1.axhline( y = over_50["thalach"].mean(),
            linestyle = "--",
            linewidth = 3)

# Figure customization
fig.suptitle(t= "Heart Disease Analysis", 
            fontsize = 20,
            fontweight = "bold")
plt.show()
```
